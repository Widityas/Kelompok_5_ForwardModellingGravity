**Forward Modelling Gravity** adalah proses **menghitung respon gaya berat (gravitational response)** dari suatu **model bawah permukaan bumi** berdasarkan distribusi densitas batuan yang diasumsikan.

Artinya, kita **menentukan bagaimana medan gravitasi seharusnya terlihat** di permukaan bumi jika kita sudah mengetahui bentuk dan sifat (densitas) dari struktur geologi di bawahnya.

Tujuannya adalah untuk **memprediksi anomali gravitasi** yang akan dihasilkan oleh model geologi tertentu.

Anomali gravitasi vertikal \( g_z \) di permukaan dapat dihitung dengan integral:

\[
g_z = G \iiint_V \frac{\rho(x, y, z)(z - z_0)}{r^3} \, dV
\]

di mana:
- \( \rho(x, y, z) \) = densitas batuan pada posisi (x, y, z)  
- \( (z - z_0) \) = jarak vertikal antara elemen massa dan titik pengamatan  
- \( r \) = jarak total antara elemen massa dan titik pengamatan  
- \( V \) = volume tubuh bawah permukaan

## ðŸ§  Kesimpulan
Forward Modelling Gravity adalah teknik **simulasi medan gravitasi** berdasarkan model densitas bawah permukaan.  
Metode ini menjadi **langkah awal penting** sebelum melakukan **inversi gravitasi** untuk memahami struktur geologi tersembunyi seperti:
- Batas lapisan sedimen
- Intrusi batuan beku
- Cekungan dan sesar
